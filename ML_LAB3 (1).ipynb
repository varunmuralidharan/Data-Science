```python
import pandas as pd
import numpy as np
Friday_data=pd.read_csv('black_friday.csv')
Friday_data
```

```python
Friday_data.isnull().sum()
```

```python
Friday_data.drop('User_ID', axis=1, inplace=True)
Friday_data.drop('Product_ID', axis=1, inplace=True)
Friday_data.drop('Product_Category_2', axis=1, inplace=True)
Friday_data.drop('Product_Category_3', axis=1, inplace=True)
```

```python
Friday_data.loc[(Friday_data['Gender'] == 'F'), 'Gender'] = 0
Friday_data.loc[(Friday_data['Gender'] == 'M'), 'Gender'] = 1
```

```python
#transforming occupation

Friday_data["Occupation"] = Friday_data["Occupation"].astype('category',Friday_data['Occupation'].unique())

# now call .get_dummies() as usual

Friday_data = pd.concat([Friday_data,pd.get_dummies(Friday_data['Occupation'], prefix='Occupation',dummy_na=True)],axis=1).drop(['Occupation'],axis=1)
```

```python
Friday_data['City_Category'].unique()
```

```python
#transforming occupation

Friday_data["City_Category"] = Friday_data["City_Category"].astype('category',categories=Friday_data['City_Category'].unique())

# now call .get_dummies() as usual
Friday_data = pd.concat([Friday_data,pd.get_dummies(Friday_data['City_Category'], prefix='City_Category',dummy_na=True)],axis=1).drop(['City_Category'],axis=1)
```

```python
Friday_data.loc[(Friday_data['Stay_In_Current_City_Years'] == '0'), 'Stay_In_Current_City_Years'] = 0
Friday_data.loc[(Friday_data['Stay_In_Current_City_Years'] == 1), 'Stay_In_Current_City_Years'] = 1
Friday_data.loc[(Friday_data['Stay_In_Current_City_Years'] == '4+'), 'Stay_In_Current_City_Years'] = 4
```

```python
#transforming occupation

Friday_data["Product_Category_1"] = Friday_data["Product_Category_1"].astype('category',categories=Friday_data['Product_Category_1'].unique())

# now call .get_dummies() as usual
Friday_data = pd.concat([Friday_data,pd.get_dummies(Friday_data['Product_Category_1'], prefix='Product_Category_1',dummy_na=True)],axis=1).drop(['Product_Category_1'],axis=1)
```

```python
Friday_data.drop('Product_Category_1_nan', axis=1, inplace=True)
Friday_data.drop('Occupation_nan', axis=1, inplace=True)
Friday_data.drop('City_Category_nan', axis=1, inplace=True)
```

```python
#split dataset in features and target variable
#feature_cols = ['Gender','Age','Occupation','City_Category','Stay_In_Current_City_Years','Marital_Status','Product_Category_1','Purchase']
#data_final_vars=Friday_data.columns.values.tolist()
K=['Purchase']
#X=[i for i in data_final_vars if i not in K]
X = Friday_data[feature_cols] # Features
y = Friday_data.Purchase # Target variable

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
np.random.seed(0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_NEW = sc.fit_transform(X_train)
scaled_features_train_x = pd.DataFrame(X_train_NEW,columns=feature_cols)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
#sc = StandardScaler()
X_test_NEW = sc.transform(X_test)
scaled_features_test_x = pd.DataFrame(X_test_NEW,columns=feature_cols)
```

```python
#Train score
from sklearn import metrics
from sklearn.metrics import r2_score
linreg = LinearRegression()

linreg.fit(scaled_features_train_x, y_train)

Y_pred_train = linreg.predict(scaled_features_train_x)

linreg.score(scaled_features_train_x, y_train)
r2_score(y_train, Y_pred_train)
#np.sqrt(metrics.mean_squared_error(y_train, Y_pred_train))
```

```python
#Test score
Y_pred_test = linreg.predict(scaled_features_test_x)

linreg.score(scaled_features_test_x, y_test)
r2_score(y_test, Y_pred_test)
#np.sqrt(metrics.mean_squared_error(y_test, Y_pred_test))
```

```python
linreg.intercept_
```

```python
from sklearn.linear_model import Ridge
alphas=[0.01, 0.1, 1, 10]

for a in alphas:
    ridge = Ridge(alpha=a, random_state = 0)
    ridge.fit(scaled_features_train_x, y_train)
    Ridge_train_score = ridge.score(scaled_features_train_x,y_train)
    Ridge_test_score = ridge.score(scaled_features_test_x, y_test)
    Ridge_train_score
    Ridge_test_score
    a
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=10, max_iter=10000,random_state=0)
lasso.fit(scaled_features_train_x, y_train)
train_score=lasso.score(scaled_features_train_x,y_train)
test_score=lasso.score(scaled_features_test_x, y_test)
coeff_used = np.sum(lasso.coef_!=0)
train_score 
test_score
coeff_used
```
